```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Import the data.
our_data = pd.read_csv("Insurance.csv")
our_data.head()
```

```python
# The iloc function allows use to select rows from the dataframe for plotting.  
x = our_data.iloc[:,:1].values
y = our_data.iloc[:,-1].values

plt.scatter(x,y,color = 'b')


plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('All Data')

plt.show()
```

```python
salary_model = LinearRegression()
```

```python
# Fit the model.
salary_model.fit(x,y)
# Plot the data and model.
y_pred = salary_model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Linear Regression')
plt.show()
```

```python
all = sns.scatterplot(x = "age", y = "charges",
            hue = "smoker", data = our_data)
all.legend(loc='lower right')
```

```python
all = sns.scatterplot(x = "age", y = "charges",
           style = "smoker", hue="bmi", data = our_data)
all.legend(loc='lower right')
```

```python
# Spliting on 3 groups, as not only age affects insurance cost and we can see 3 groups/3 lines of dots
# on the chart above.

# First one: no smoker.

# The iloc function allows use to select rows from the dataframe for plotting. 

our_data_no_smoker = our_data[our_data["smoker"]== "no"]

x1 = our_data_no_smoker.iloc[:,:1].values
y1 = our_data_no_smoker.iloc[:,-1].values

plt.scatter(x1,y1,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('No smoker')
plt.show()
```

```python
# Plot the data and model.
salary_model.fit(x1,y1)
y_pred1 = salary_model.predict(x1)

plt.scatter(x1,y1,color = 'b')
plt.plot(x1,y_pred1,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Linear Regression for No smoker')
plt.show()
```

```python
# Predict an unknown value.
unk_x = [[70]] 

x_pred = np.append(x1, unk_x).reshape(-1,1)
y_pred = salary_model.predict(x_pred)

plt.scatter(x1,y1,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Predict an unknown value for No smoker')

plt.show()

print("Charges should be:", salary_model.predict(unk_x))
```

```python
# Spliting on 3 groups.

# Second group: smoker, bmi >30.

# If BMI is 30.0 or higher, it falls within the obese range.
# BMI is a useful measure of overweight and obesity. It is calculated from your height and weight.
# BMI is an estimate of body fat and a good gauge of your risk for diseases that can occur with more body fat.


# The iloc function allows use to select rows from the dataframe for plotting. 

our_data_smoker = our_data[our_data["smoker"]=="yes"]
our_data_smoker_bmi_high =our_data_smoker[our_data_smoker["bmi"]>=30]

x2 = our_data_smoker_bmi_high.iloc[:,:1].values
y2 = our_data_smoker_bmi_high.iloc[:,-1].values

plt.scatter(x2,y2,color = 'g')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Smoker and obese')
plt.show()
```

```python
# Plot the data and model.
salary_model.fit(x2,y2)

y_pred2 = salary_model.predict(x2)

plt.scatter(x2,y2,color = 'g')
plt.plot(x2,y_pred2,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Linear Regression for smoker and obese')
plt.show()
```

```python
# Spliting on 3 groups.

# Third one: smoker, bmi <= 30.

# The iloc function allows use to select rows from the dataframe for plotting. 

our_data_smoker = our_data[our_data["smoker"]=="yes"]
our_data_smoker_bmi =our_data_smoker[our_data_smoker["bmi"] < 30]

x3 = our_data_smoker_bmi.iloc[:,:1].values
y3 = our_data_smoker_bmi.iloc[:,-1].values

plt.scatter(x3,y3,color = 'aqua')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Smoker and normal bmi')
plt.show()
```

```python
# Plot the data and model.
salary_model.fit(x3,y3)

y_pred3 = salary_model.predict(x3)

plt.scatter(x3,y3,color = 'aqua')
plt.plot(x3,y_pred3,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Linear Regression for smoker and normal bmi')
plt.show()
```
